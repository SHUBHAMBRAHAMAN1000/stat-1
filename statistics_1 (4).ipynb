```python
# @title 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt

# Number of samples and range of uniform distribution
n = 1000  # Number of samples
low = 1  # Minimum value
high = 6  # Maximum value

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(low, high+1, size=n)

# Calculate the frequency of each outcome
values, counts = np.unique(samples, return_counts=True)

# Calculate the Probability Mass Function (PMF)
pmf = counts / n

# Plot PMF
plt.bar(values, pmf, width=0.1)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')
plt.show()
```

```python
# @title 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    """Calculate the PDF of a Bernoulli distribution"""
    # The two possible outcomes: 0 and 1
    outcomes = [0, 1]
    pdf = [1 - p, p]  # Probability of failure (0) and success (1)

    return outcomes, pdf

# Parameters
p = 0.6  # Probability of success

# Get the PDF
outcomes, pdf = bernoulli_pdf(p)

# Plot the PDF
plt.bar(outcomes, pdf, width=0.1)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution (p = {p})')
plt.show()
```

```python
# @title 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

# Number of samples and range for the uniform distribution
n = 1000  # Number of samples
low = 1  # Minimum value
high = 6  # Maximum value

# Generate random samples from a discrete uniform distribution
samples = np.random.randint(low, high + 1, size=n)

# Calculate the CDF
values, counts = np.unique(samples, return_counts=True)
cdf = np.cumsum(counts) / n

# Plot the CDF
plt.step(values, cdf, where='post', color='blue', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.show()
```

```python
# @title 9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """Calculate the Z-scores for a given dataset"""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Sample data (can be any dataset)
data = np.random.normal(0, 1, 1000)  # Normally distributed data with mean=0, std_dev=1

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Z-scores of a Normally Distributed Dataset')
plt.show()
```

```python
# @title 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate the Z-score
def calculate_z_score(x, mean, std_dev):
    return (x - mean) / std_dev

# Example dataset (Normal distribution)
data = np.random.normal(0, 1, 1000)  # Mean=0, Std=1

# Calculate mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Example data point
x = 2

# Calculate the Z-score
z_score = calculate_z_score(x, mean, std_dev)

# Plot the standard normal distribution for comparison
x_vals = np.linspace(-5, 5, 1000)
y_vals = norm.pdf(x_vals, loc=0, scale=1)

plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')
plt.axvline(x=z_score, color='red', linestyle='dashed', label=f'Z-score: {z_score:.2f}')
plt.title('Z-score Comparison with Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print(f"Z-score for data point {x} is: {z_score:.2f}")
```

```python
# @title 15.Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Function to perform a Z-test
def z_test(sample, population_mean, population_std, sample_size):
    sample_mean = np.mean(sample)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    return z_statistic

# Parameters
population_mean = 50  # Population mean
population_std = 10   # Population standard deviation
sample_size = 30      # Sample size

# Generate a random sample from a normal distribution
sample = np.random.normal(population_mean, population_std, sample_size)

# Perform Z-test
z_stat = z_test(sample, population_mean, population_std, sample_size)

# Calculate the p-value for a two-tailed test (using standard normal distribution)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

```

```python
# @title 17.Generate random variables and calculate their corresponding probabilities using the binomial distribution.
from scipy.stats import binom

def binomial_distribution(n, p, size):
    # Generate random variables following a binomial distribution
    random_variables = binom.rvs(n, p, size=size)

    # Calculate the probability mass function (PMF) for the random variables
    probabilities = binom.pmf(random_variables, n, p)

    return random_variables, probabilities

# Example usage: n = 10, p = 0.5, size = 1000
n = 10
p = 0.5
size = 1000
random_vars, probabilities = binomial_distribution(n, p, size)

# Output the first 10 random variables and their corresponding probabilities
print("Random Variables (First 10):", random_vars[:10])
print("Corresponding Probabilities (First 10):", probabilities[:10])
```

```python
# @title 18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    # Z-score formula: (data_point - mean) / std_dev
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    # Calculate the probability of observing a Z-score <= given Z using the CDF of the standard normal distribution
    probability = norm.cdf(z_score)
    return probability

# Example usage:
data_point = 2  # example data point
mean = 0        # mean of the standard normal distribution
std_dev = 1     # standard deviation of the standard normal distribution

z_score = calculate_z_score(data_point, mean, std_dev)
probability = compare_to_standard_normal(z_score)

print(f"Z-score: {z_score}")
print(f"Probability of Z <= {z_score}: {probability}")

```

```python
# @title 19.Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, sample_size, sample_std_dev):
    # Calculate the Z-statistic: (sample_mean - population_mean) / (sample_std_dev / sqrt(sample_size))
    z_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

    return z_statistic

def hypothesis_test(z_statistic, alpha=0.05):
    # Compare the Z-statistic to the critical value at a significance level of alpha (e.g., 0.05)
    critical_value = norm.ppf(1 - alpha / 2)

    # If the absolute Z-statistic is greater than the critical value, reject the null hypothesis
    if abs(z_statistic) > critical_value:
        return "Reject the null hypothesis"
    else:
        return "Fail to reject the null hypothesis"

# Example usage:
sample_mean = 105
population_mean = 100
sample_size = 30
sample_std_dev = 15

z_statistic = z_test(sample_mean, population_mean, sample_size, sample_std_dev)
result = hypothesis_test(z_statistic)

print(f"Z-statistic: {z_statistic}")
print("Hypothesis test result:", result)
```

```python
# @title 20.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    # Calculate sample mean and sample standard deviation
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation
    n = len(data)

    # Calculate the standard error of the mean
    standard_error = std_dev / np.sqrt(n)

    # Find the z-score for the given confidence level
    z_score = stats.norm.ppf((1 + confidence_level) / 2)

    # Calculate the margin of error
    margin_of_error = z_score * standard_error

    # Calculate the confidence interval
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage:
data = np.random.normal(loc=100, scale=15, size=30)  # Generate a sample dataset
confidence_interval_result = confidence_interval(data)

print(f"Confidence Interval: {confidence_interval_result}")
```

```python
# @title 21.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

def confidence_interval_for_mean(data, confidence_level=0.95):
    # Calculate the sample mean and sample standard deviation
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation
    n = len(data)

    # Calculate the standard error of the mean
    standard_error = std_dev / np.sqrt(n)

    # Find the t-score for the given confidence level
    t_score = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

    # Calculate the margin of error
    margin_of_error = t_score * standard_error

    # Calculate the confidence interval
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage:
data = np.random.normal(loc=100, scale=15, size=30)  # Generate data from a normal distribution
confidence_interval_result = confidence_interval_for_mean(data)

print(f"Confidence Interval for the Mean: {confidence_interval_result}")
```

```python
# @title 23.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(mu, x_range):
    # Calculate the CDF of the Poisson distribution
    cdf = poisson.cdf(x_range, mu)

    # Plot the CDF
    plt.plot(x_range, cdf, label=f"Poisson Distribution (mu={mu})")
    plt.title("Cumulative Distribution Function (CDF) of a Poisson Distribution")
    plt.xlabel("X")
    plt.ylabel("Cumulative Probability")
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage:
mu = 5  # Mean of the Poisson distribution
x_range = np.arange(0, 15)  # X values (0 to 14)
plot_poisson_cdf(mu, x_range)
```

```python
# @title 26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np

def calculate_range_and_iqr(data):
    # Calculate the range (max - min)
    data_range = np.max(data) - np.min(data)

    # Calculate the interquartile range (IQR)
    Q1 = np.percentile(data, 25)  # First quartile (25th percentile)
    Q3 = np.percentile(data, 75)  # Third quartile (75th percentile)
    iqr = Q3 - Q1

    return data_range, iqr

# Example usage:
data = np.random.normal(loc=100, scale=15, size=50)  # Generate a sample dataset

data_range, iqr = calculate_range_and_iqr(data)

print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {iqr}")
```

```python
# @title 27.Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)

    # Apply Z-score normalization
    normalized_data = (data - mean) / std_dev

    return normalized_data

def plot_data_before_and_after_normalization(original_data, normalized_data):
    # Plot original data and normalized data
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=15, color='blue', alpha=0.7)
    plt.title("Original Data Distribution")

    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=15, color='green', alpha=0.7)
    plt.title("Normalized Data Distribution")

    plt.show()

# Example usage:
data = np.random.normal(loc=100, scale=15, size=50)  # Generate a sample dataset

normalized_data = z_score_normalization(data)

# Visualize the transformation
plot_data_before_and_after_normalization(data, normalized_data)

# Print the first few values of original and normalized data
print("Original Data (First 5):", data[:5])
print("Normalized Data (First 5):", normalized_data[:5])
```

```python
# @title 28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    # Calculate the skewness and kurtosis of the dataset
    skewness = skew(data)
    kurt = kurtosis(data)

    return skewness, kurt

# Example usage:
data = np.random.normal(loc=100, scale=15, size=1000)  # Generate data from a normal distribution

skewness, kurt = calculate_skewness_and_kurtosis(data)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```
